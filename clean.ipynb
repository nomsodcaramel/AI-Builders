```python
import pandas as pd

# Read in the data
df = pd.read_csv('Python_Exercises_final.csv')

df.head(20)
```

```python
df['Merge'] = df['Description'] + "\n" + df['Code']
print(df['Merge'][0])
```

```python
df["Test_Case"] = df["Test_Case"].str.replace("From code\n", "\nOutput : ")
print(df["Test_Case"].head(20))
```

```python
df["Test_Case"] = df["Test_Case"].str.replace("Output :", "#Output : ")
df["Test_Case"] = df["Test_Case"].str.replace("Input :", "#Input : ")
```

```python
print(df["Test_Case"])
```

```python
df['Merge'] = df['Merge'] + df['Test_Case'] + "\n[End]"
print(df['Merge'][0])
```

```python
df.to_csv("Python_Exercises_final.csv", index=False)
```
